```python
import pandas as pd
import os
import geopandas as gpd
```

```python
balita_df = pd.read_csv('../datasets/merged_dataset.csv')
balita_df.head()
```

```python
balita_df.columns
```

```python
cols = ['nama_kabupaten_kota','tahun','jumlah_anak_usia_0-23_bulan',
       'jumlah_anak_0-23_terdaftar_jaminan_kesehatan',
       'jumlah_anak_0-23_pertumbuhan_hijau',
       'jumlah_anak_0-23_pertumbuhan_kuning',
       'jumlah_anak_0-23_pertumbuhan_merah',
       'jumlah_balita_stunting', 'persentase_balita_stunting',
       'penemuan_kasus_pneumonia', 'kasus_pneumonia_berat',
       'jumlah_kematian_kasus_pneumonia', 'jumlah_perkiraan_kasus_pneumonia',
       'jumlah_balita_gizi_kurang_2019', 'jumlah_balita_kurus_2019',
       'jumlah_balita_pendek_2019','jumlah_balita_gizi_kurang']
```

```python
balita_df_filter = balita_df[cols]
balita_df_filter.head()
```

```python
balita_df_filter['kab/kota'] = balita_df_filter['nama_kabupaten_kota'].apply(lambda x: x.title().replace('Kabupaten','').strip())
balita_df_filter.head()
```

```python
balita_df_filter.to_csv('../datasets/ready/balita_data.csv',index=False)
```

### Balita Stunting

```python
balita_stunting = balita_df_filter[['nama_kabupaten_kota','kab/kota','tahun']].copy()
balita_stunting['keterangan'] = 'Balita Stunting'
balita_stunting['total'] = balita_df_filter['jumlah_balita_stunting']
display(balita_stunting.head())
balita_stunting.to_csv('../datasets/ready/balita_stunting.csv',index=False)
```

### Balita 0-23 bulan

```python
balita_pertumbuhan = balita_df_filter[['nama_kabupaten_kota', 'kab/kota', 'tahun', 'jumlah_anak_usia_0-23_bulan', 'jumlah_anak_0-23_terdaftar_jaminan_kesehatan', 'jumlah_anak_0-23_pertumbuhan_hijau', 'jumlah_anak_0-23_pertumbuhan_kuning', 'jumlah_anak_0-23_pertumbuhan_merah']].copy()
balita_pertumbuhan.tail()
```

```python
balita_pertumbuhan_total = balita_pertumbuhan[['nama_kabupaten_kota','tahun','kab/kota']].copy()
balita_pertumbuhan_total['keterangan'] = 'Total Anak 0-23 Bulan'
balita_pertumbuhan_total['total'] = balita_pertumbuhan['jumlah_anak_usia_0-23_bulan'].copy()

balita_pertumbuhan_hijau = balita_pertumbuhan[['nama_kabupaten_kota','tahun','kab/kota']].copy()
balita_pertumbuhan_hijau['keterangan'] = 'Pertumbuhan Hijau'
balita_pertumbuhan_hijau['total'] = balita_pertumbuhan['jumlah_anak_0-23_pertumbuhan_hijau'].copy()
balita_pertumbuhan_hijau

balita_pertumbuhan_kuning = balita_pertumbuhan[['nama_kabupaten_kota','tahun','kab/kota']].copy()
balita_pertumbuhan_kuning['keterangan'] = 'Pertumbuhan Kuning'
balita_pertumbuhan_kuning['total'] = balita_pertumbuhan['jumlah_anak_0-23_pertumbuhan_kuning'].copy()

balita_pertumbuhan_merah = balita_pertumbuhan[['nama_kabupaten_kota','tahun','kab/kota']].copy()
balita_pertumbuhan_merah['keterangan'] = 'Pertumbuhan Merah'
balita_pertumbuhan_merah['total'] = balita_pertumbuhan['jumlah_anak_0-23_pertumbuhan_merah'].copy()
balita_pertumbuhan_merah
```

```python
balita_0_23 = pd.concat([balita_pertumbuhan_total, balita_pertumbuhan_hijau, balita_pertumbuhan_kuning, balita_pertumbuhan_merah]).reset_index(drop=True)
balita_0_23
```

```python
balita_0_23.to_csv('../datasets/ready/balita_0-23_bulan.csv',index=False)
```

### Balita Pneumonia

```python
balita_pneumonia = balita_df_filter[['nama_kabupaten_kota','kab/kota','tahun','jumlah_balita_stunting']].copy()


balita_pneumonia_kasus = balita_pneumonia[['nama_kabupaten_kota','tahun','kab/kota']].copy()
balita_pneumonia_kasus['keterangan'] = 'Balita Stunting'
balita_pneumonia_kasus['total'] = balita_pneumonia['jumlah_balita_stunting'].copy()
balita_pneumonia_kasus
```

```python
balita_pneumonia = balita_df_filter[['nama_kabupaten_kota','kab/kota','tahun','penemuan_kasus_pneumonia', 
                                     'kasus_pneumonia_berat','jumlah_kematian_kasus_pneumonia', 
                                     'jumlah_perkiraan_kasus_pneumonia']].copy()


balita_pneumonia_kasus = balita_pneumonia[['nama_kabupaten_kota','tahun','kab/kota']].copy()
balita_pneumonia_kasus['keterangan'] = 'Penemuan Kasus Pneumonia'
balita_pneumonia_kasus['total'] = balita_pneumonia['penemuan_kasus_pneumonia'].copy()

balita_pneumonia_berat = balita_pneumonia[['nama_kabupaten_kota','tahun','kab/kota']].copy()
balita_pneumonia_berat['keterangan'] = 'Kasus Pneumonia Berat'
balita_pneumonia_berat['total'] = balita_pneumonia['kasus_pneumonia_berat'].copy()

balita_pneumonia_mati = balita_pneumonia[['nama_kabupaten_kota','tahun','kab/kota']].copy()
balita_pneumonia_mati['keterangan'] = 'Kematian Akibat Pneumonia'
balita_pneumonia_mati['total'] = balita_pneumonia['jumlah_kematian_kasus_pneumonia'].copy()

balita_pneumonia_kira = balita_pneumonia[['nama_kabupaten_kota','tahun','kab/kota']].copy()
balita_pneumonia_kira['keterangan'] = 'Perkiraan Kasus Pneumonia'
balita_pneumonia_kira['total'] = balita_pneumonia['jumlah_perkiraan_kasus_pneumonia'].copy()

balita_pneumonia_new = pd.concat([balita_pneumonia_kasus, balita_pneumonia_berat, balita_pneumonia_mati, balita_pneumonia_kira]).reset_index(drop=True)
balita_pneumonia_new
```

```python
balita_pneumonia_new.to_csv('../datasets/ready/balita_pneumonia.csv',index=False)
```
